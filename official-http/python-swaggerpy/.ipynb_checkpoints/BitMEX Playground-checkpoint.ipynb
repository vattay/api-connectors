```python
from bravado.client import SwaggerClient
from bravado.requests_client import RequestsClient
from BitMEXAPIKeyAuthenticator import APIKeyAuthenticator

HOST = "https://bitmex.com"
SPEC_URI = HOST + "/api/explorer/swagger.json"

# See full config options at http://bravado.readthedocs.io/en/latest/configuration.html
config = {
  # Don't use models (Python classes) instead of dicts for #/definitions/{models}
  'use_models': False,
  # This library has some issues with nullable fields
  'validate_responses': False,
}

bitMEX = SwaggerClient.from_url(
  SPEC_URI,
  config=config)


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from datetime import datetime
from datetime import timedelta

utcnow = datetime.utcnow()
lookback = timedelta(hours=1)
start_time = (utcnow - lookback)

pageSize = 500
start_at = 0
xbtusd = []
result = bitMEX.Trade.Trade_get(symbol='XBTUSD', count=500, reverse=True, startTime=start_time).result();
while len(result) > 0:
    xbtusd += result
    start_at += pageSize
    result = bitMEX.Trade.Trade_get(symbol='XBTUSD', count=500, reverse=True, startTime=start_time, start=start_at).result();


bxbt = bitMEX.Trade.Trade_get(symbol='.BXBT', count=500, reverse=True).result();
```

```python
df_xbtusd = pd.DataFrame.from_records(xbtusd, index='timestamp')
s_xbtusd_1min = pd.DataFrame.from_records(xbtusd, index='timestamp').price.resample('1min').ohlc()
s_xbtusd_vol = pd.DataFrame.from_records(xbtusd, index='timestamp')['size'].resample('1min').mean()
s_xbtusd_1sec = pd.DataFrame.from_records(xbtusd, index='timestamp').price.resample('1s').ohlc()
df_bxbt = pd.DataFrame.from_records(bxbt, index='timestamp')
```

```python
flipflow = {}
last_side = None
change = 0.0
avg_change = 0.0
pivot_price = 0.0
cumulative_vol = 0.0
for i, row in df_xbtusd.iterrows():
    if last_side is None:
        last_side = row.side
        pivot_price = row.price
    elif last_side != row.side:
        if row.price != pivot_price:
            cumulative_vol += row['size']
            if cumulative_vol > 5000:
                flipflow[i] = abs(row.price - pivot_price)
                last_side = row.side
                pivot_price = row.price
                cumulative_vol = 0.0
df_flipflow = pd.Series(flipflow)
print df_flipflow.mean()
print df_flipflow.max()
print df_flipflow.min()
print df_flipflow.std()

# print (df_flipflow.sum() / len(df_flipflow))** 0.5

print df_flipflow.hist(bins=20)
```
